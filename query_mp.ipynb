# Querying for full dataset

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('labels.csv') 
```

```python
ids = df1['Identifier'].values.tolist()
```

Full dataset too big for MP limit (10k) - will split data in two

```python
ids_1 = ids[:9000]
ids_2 = ids[9000:]
print(len(ids), len(ids_1)+len(ids_2))
```

```python
with MPRester("GdU0AaMi4XS8vSb2YOsMRRyEgnD3quCD") as mpr:
    docs = mpr.materials.summary.search(material_ids=ids_1, fields = ["material_id","formula_pretty", "energy_above_hull", "structure"])
```

```python
docs_1 = docs
```

```python
with MPRester("GdU0AaMi4XS8vSb2YOsMRRyEgnD3quCD") as mpr:
    docs_2 = mpr.materials.summary.search(material_ids=ids_2, fields = ["material_id","formula_pretty", "energy_above_hull", "structure"])
```

```python
print(len(docs_1),len(docs_2),len(docs_1)+len(docs_2))
```

```python
docs=docs_1+docs_2
```

```python
structures = []
formulae = []
eah =[]
ids = []
for doc in docs:
    structures.append(JAA.get_atoms(doc.structure))
    formulae.append(doc.formula_pretty)
    eah.append(doc.energy_above_hull)
    ids.append(doc.material_id)
```

```python
print("now make a dataframe and join to labels")
```
