```python
from jarvis.core.atoms import Atoms
from jarvis.db.jsonutils import dumpjson
from jarvis.db.jsonutils import loadjson
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Yu_Zunger_best.csv')
```

```python
input = []
test_dict = {}
for icsd in df.ICSD:
    try:
        ats = Atoms.from_cif('best_cifs/'+str(icsd)+'.cif',use_cif2cell=False)
        slme = float(df.loc[df.ICSD==icsd].SLME)
        test_dict[icsd] = slme
        d = {'jid': icsd,'atoms':ats.to_dict(),'target':slme}
        input.append(d)
    except:
        print(f"Structure {icsd} failed. Not added to input list")
```

```python
len(input)
```

```python
dumpjson(input,'Yu_test.json')
```

```python
dumpjson(test_dict,'Yu_test_dict.json')
```
