# Generating test set

```python
import pandas as pd
import numpy as np
from jarvis.db.jsonutils import dumpjson
from jarvis.db.jsonutils import loadjson
import math
```

```python
alphas = loadjson('../mpcontribs-downloads/100_alphas_unscaled.json')
```

```python
snums = loadjson('../ch4/mp_all_snumat_data.json')
```

```python
ft_ids = list(alphas.keys())
```

```python
fab_data = loadjson('../hybrid_gaps/clean_fabs.json')
```

```python
fab_ids = [mat['jid'] for mat in fab_data]
```

```python
ids = set(snums).intersection(ft_ids)
```

```python
test_ids = [i for i in fab_ids] # make copy of fab_ids that will be appended to with leave-out test set
```

```python
test_size=0.1*len(ids)
```

```python
indices = np.arange(0,len(ids),10)
```

```python
len(indices)
```

```python
indices = list(indices)
```

```python
for i in indices:
    test_ids.append(list(ids)[i])
```

```python
len(test_ids)
```

```python
with open('test_ids.txt','w+') as file:
    for id in test_ids:
        file.write(id+'\n')
```

```python
form_struc = loadjson('form_struc.json')
```

```python
fab_strucs = loadjson('fab_structures.json')
```

```python
len(fab_strucs)
```

```python
fs = {}
for i in fab_strucs:
    fs[i['jid']] = i['atoms']
```

```python
input = []
for tid in test_ids:
    if tid in form_struc:
        input.append({'jid':tid,'atoms':form_struc[tid]})
    elif tid in fs:
        input.append({'jid':tid,'atoms':fs[tid]})
    else:
        print("Structure missing for MPID: ",tid)
```

```python
dumpjson(input,'test_input.json')
```
