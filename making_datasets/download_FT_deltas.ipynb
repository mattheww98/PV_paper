```python
fun_gaps = {}
dir_gaps = {}
da_gaps = {}
deltas = {}
edge_deltas = {}
FT_gap_data = {}
for d in data['data']:
    mpid = d['identifier']
    fun_gap = d['data']['properties']['optical']['bandgaps']['GGA']['value']
    dir_gap = d['data']['properties']['optical']['bandgaps']['GGA|d']['value']
    da_gap = d['data']['properties']['optical']['bandgaps']['GGA|da']['value']
    delta = da_gap - fun_gap
    edge_delt = d['data']['properties']['optical']['delta|edge']['GGA|d']['value']
    fun_gaps[mpid] = fun_gap
    dir_gaps[mpid] = dir_gap
    da_gaps[mpid] = da_gap
    deltas[mpid] = delta
    edge_deltas[mpid] = edge_delt
    FT_gap_data[mpid] = {"fun_gap": fun_gap, "dir_gap": dir_gap, "da_gap": da_gap, "offset": delta}
```
