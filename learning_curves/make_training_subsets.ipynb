# Making training sets of different sizes for learning curves

```python
from jarvis.db.jsonutils import dumpjson
from jarvis.db.jsonutils import loadjson
```

```python
def make_subsets(fname,prop):
    data = loadjson("../"+fname+".json")
    for i in [10,100,1000]:
        for j in range(4):
            index = 1000*j+i
            subset = data[1000*j:index]
            outname = prop+"/"+prop+"_"+str(index)+".json"
            dumpjson(subset,outname)
        
```

```python
make_subsets("FT_etas_input","etas")
```

```python
make_subsets("corrs_input","corr")
```

```python
make_subsets("As_input","As")
```

```python
make_subsets("alphas_input","alphas")
```

```python
make_subsets("FT_deltas_input","deltas")
```

```python
make_subsets("gaps_input","gaps")
```
